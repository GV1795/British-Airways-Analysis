```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("D:/GAURAV NOTES/notes 1965/DataSet/reviews_data1.csv/reviews_data1.csv")
df.head()
```

```python
# Basic data exploration
print("\nData Exploration:")
print(df.info())
```

```python
# Count the occurrences of each unique value in the 'Verified' column
verified_counts = df['Verified'].value_counts()
df
```

```python
# Count the occurrences of each unique value in the 'Recommended' column
recommended_counts = df['Recommended'].value_counts()
df
```

```python
# Display the counts
print("\nVerified Counts:")
print(verified_counts)

print("\nRecommended Counts:")
print(recommended_counts)
```

```python
# Visualization: Bar chart for 'Verified' counts
plt.figure(figsize=(8, 6))
plt.bar(verified_counts.index, verified_counts.values, color=['green', 'red'])
plt.xlabel('Verified Status')
plt.ylabel('Number of Reviews')
plt.title('Distribution of Verified Reviews')
plt.show()
```

```python
# Visualization: Pie chart for 'Recommended' counts
plt.figure(figsize=(8, 8))
plt.pie(recommended_counts, labels=recommended_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Recommended Reviews')
plt.show()
```

```python
# Read the CSV file into a DataFrame
df1 = pd.read_csv("D:/GAURAV NOTES/notes 1965/DataSet/rating_data.csv")
df1.head()
```

```python
# Basic data exploration
print("\nData Exploration:")
print(df.info())
```

```python
# Count the occurrences of each unique value in the 'Aircraft' column
aircraft_counts = df1['Aircraft'].value_counts()
# Count the occurrences of each unique value in the 'Type Of Traveler' column
traveler_counts = df1['Type Of Traveller'].value_counts()
```

```python
# Display the counts
print("\nAircraft Counts:")
print(aircraft_counts)

print("\nType Of Traveller Counts:")
print(traveler_counts)
```

```python
# Get the top 5 aircraft types
top_aircraft = aircraft_counts.head(5)

# Visualization: Bar chart for top 5 'Aircraft' counts
plt.figure(figsize=(15, 8))  # Adjust the figure size as needed
plt.bar(top_aircraft.index, top_aircraft.values, color='skyblue')
plt.xlabel('Aircraft Type')
plt.ylabel('Number of Ratings')
plt.title('Top 5 Aircraft Types by Number of Ratings')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()

```

```python
# Visualization: Pie chart for 'Type Of Traveler' counts
plt.figure(figsize=(8, 8))
plt.pie(traveler_counts, labels=traveler_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ratings by Type Of Traveller')
plt.show()
```

```python
# Count the occurrences of each unique value in the 'Aircraft' column for each 'Seat Type'
top_aircraft_by_seat_type = df1.groupby('Seat Type')['Aircraft'].value_counts().groupby(level=0, group_keys=False).nlargest(5)
df1.head()
```

```python
# Calculate the average ratings for each category
average_ratings = df1[['Seat Comfort', 'Cabin Staff Service', 
                      'Food & Beverages', 'Inflight Entertainment', 
                      'Ground Service', 'Wifi & Connectivity', 'Value For Money']].mean()
# Display average ratings
print("Average Ratings:")
print(average_ratings)
```

```python
# Plotting
# Calculate the average ratings for each category
average_ratings = df1[['Seat Comfort', 'Cabin Staff Service', 
                      'Food & Beverages', 'Inflight Entertainment', 
                      'Ground Service', 'Wifi & Connectivity', 'Value For Money']].mean()
# Display average ratings
print("Average Ratings:")
print(average_ratings)
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
average_ratings.plot(kind='bar', color='lightblue')
plt.title('Average Ratings for Flight Categories')
plt.xlabel('Categories')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```
